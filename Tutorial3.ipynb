```python
import numpy as np

array_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 array:\n", array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print("Random 3x4 array:\n", random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))

print("Array of ones:\n", ones_array)
print("Array of zeroes:\n", zeros_array)
```

```python
random_6x3 = np.random.rand(6, 3)
transposed_array = np.transpose(random_6x3)

print("Original 6x3 array:\n", random_6x3)
print("\nTransposed array (3x6):\n", transposed_array)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)

dot_product = np.dot(A, B)

print("Matrix A (n x p):\n", A)
print("\nMatrix B (p x m):\n", B)
print("\nDot product (n x m):\n", dot_product)
```
